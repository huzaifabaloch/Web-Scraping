```python
from bs4 import BeautifulSoup as soup
from requests import get
```

```python
# the url or page from where to scrap the data.
url = "http://www.espncricinfo.com/rankings/content/page/211271.html"
```

```python
response = get(url)  # checking the response from the server to get the entire page.
```

```python
html_soup = soup(response.text, 'html.parser')  # to structure the entire code in html format using beautiful soup.
```

```python
main_heading = html_soup.find_all('h1')[1].text
```

```python
rankings = html_soup.find_all('table', {'class':'StoryengineTable'})
formate = html_soup.find_all('div', {'class':'ciPhotoContainer'})[0].find_all('h3')
rankings[0]
```

```python
for i, r in enumerate(rankings):
    print(formate[i].text)
    print(rankings[0].caption.text)

    
    tr = r.find_all('tr')
    
    i=0
    
    for t in tr:
        if i == 0:
            search = 'th'
        else:
            search = 'tr'
            
        for c in t.findChildren(search):
            print('{0:20}'.format(c.text), end=' ')
        print()
        i += 1
    print("\n\n")
    
```
